## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
# Read the AirPassengers dataset
airline = pd.read_csv('/content/traffic.csv',
                       index_col ='DateTime',
                      parse_dates = True)
```

## show dataset

```python
# Print the first five rows of the dataset
airline.head()
```

## Apply ETS

Try 'additive'

```python
# ETS Decomposition
result = seasonal_decompose(airline['Vehicles'],period = 100,
                            model ='multiplicative')
```

## Plot ETS

```python
# ETS plot
result.plot()
```
